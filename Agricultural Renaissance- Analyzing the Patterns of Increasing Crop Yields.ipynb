## About this dataset

This dataset provides data on crop yields, harvested areas, and production quantities for wheat, maize, rice, and soybeans. Crop yields are the harvested production per unit of harvested area for crop products. In most cases yield data are not recorded but are obtained by dividing the production data by the data on the area harvested. The actual yield that is captured on a farm depends on several factors such as the crop's genetic potential, the amount of sunlight, water, and nutrients absorbed by the crop, the presence of weeds and pests. This indicator is presented for wheat, maize, rice, and soybean. Crop production is measured in tonnes per hectare.

This dataset includes information on crop production from 2010-2016

How to use the dataset

https://www.kaggle.com/usda/crop-production

Crop production is an important economic activity that affects commodity prices and macroeconomic uncertainty. This dataset provides data on crop yields, harvested areas, and production quantities for wheat, maize, rice, and soybeans. The data are presented in tonnes per hectare, in thousand hectares, and in thousand tonnes.

## About Indicators

* LOCATION: The country or region where the crop is grown. (String)

### Theorem
### Solve missing
* Ostrict theorem - remove the whole row if a data is missing from it. (Not recommended)
* Mean imputation - if any values are missing we take mean of all the values and fill the missing values with the mean values. (Used on field with loose requirements for accurate data. ie. not used in sensitive data example for medical practices)

```python
# Replace the ''(null) values with NaN
df = df.replace(" ",np.NaN)          # if there is missing values

# Impute the missing values with mean imputation
df = df.fillna(df.mean())                         # Mean Theorem

# Count the number of NaNs in the dataset to verify
print(df.isnull().values.sum())
```

```python
# Removing the column having entire missing value
df = df.dropna(axis = 1, how ='all')
```

```python
# Reinspecting the missing value in dataset
print(df.isnull().values.sum())
```

```python
sns.set_style('whitegrid')

# Group the data by 'TIME' and calculate the mean yield for each year
grouped_data = df.groupby('TIME')['Value'].mean().reset_index()

# Plot the time series of average crop yields
plt.figure(figsize=(14, 7))
sns.lineplot(data=grouped_data, x='TIME', y='Value', marker='o')
plt.title('Average Crop Yield Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Yield (tonnes per hectare)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Interpretation
* The above line graphs shows the yeild of crops over the year.
* There is a general upward trend in average crop yields over the years.
* The overall trend is positive, indicating that on average, crop production has become more efficient or that conditions have been increasingly favorable for crop growth.
* Between the year 1993 and 1995 there is exponential growth in the production of crops.

### Different data can be further analysed to find out: 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming we have a dataframe 'climate_data' with 'Year', 'Temperature', and 'Crop Yield' columns
# This is a placeholder for the actual climate data which needs to be provided

# Create a sample dataframe to simulate the climate data
# This is just for demonstration purposes
climate_data = pd.DataFrame({
    'Year': pd.date_range(start='1990', periods=30, freq='Y').year,
    'Temperature': np.random.normal(loc=15, scale=5, size=30), # Random normal distribution
    'Crop Yield': np.random.normal(loc=2, scale=0.5, size=30) # Random normal distribution
})

# Visualize the relationship between temperature and crop yields
plt.figure(figsize=(14, 7))
sns.lineplot(data=climate_data, x='Year', y='Temperature', marker='o', label='Temperature')
sns.lineplot(data=climate_data, x='Year', y='Crop Yield', marker='o', label='Crop Yield')
plt.title('Temperature and Crop Yield Over Time')
plt.xlabel('Year')
plt.ylabel('Temperature / Crop Yield')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create sample data for precipitation and CO2 levels
climate_data['Precipitation'] = np.random.normal(loc=100, scale=30, size=30) # Random normal distribution
climate_data['CO2 Levels'] = np.random.normal(loc=400, scale=30, size=30) # Random normal distribution

# Visualize the relationship between precipitation and crop yields
plt.figure(figsize=(14, 7))
sns.lineplot(data=climate_data, x='Year', y='Precipitation', marker='o', label='Precipitation')
sns.lineplot(data=climate_data, x='Year', y='Crop Yield', marker='o', label='Crop Yield')
plt.title('Precipitation and Crop Yield Over Time')
plt.xlabel('Year')
plt.ylabel('Precipitation / Crop Yield')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualize the relationship between CO2 levels and crop yields
plt.figure(figsize=(14, 7))
sns.lineplot(data=climate_data, x='Year', y='CO2 Levels', marker='o', label='CO2 Levels')
sns.lineplot(data=climate_data, x='Year', y='Crop Yield', marker='o', label='Crop Yield')
plt.title('CO2 Levels and Crop Yield Over Time')
plt.xlabel('Year')
plt.ylabel('CO2 Levels / Crop Yield')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
